## Trying to get a smaller version of the dataset, keeping only important columns

```python
df[["Area", "Item", "Element", "Unit"]]
```

```python
# get a subset of columns that start with "Y" and do not end in "F" or "N"
df.filter(regex="^Y(?!.*[FN])")
```

```python
# get a slimmer dataset
df_s = df[["Area", "Item", "Element", "Unit"]].join(
    df.filter(regex="^Y(?!.*[FN])", axis=1)
)
df_s
```

```python
df_almond = df_s[df_s.Item.str.contains("Almond", case=False, na=False)]
df_almond.shape
```

```python
df_almond
```

```python
# Get a list of countries that produce Almonds
df_almond.Area.unique(), df_almond.Area.nunique()
```

```python
# Get a list of Elements about Amonds production
df_almond.Element.unique(), df_almond.Element.nunique()
```

```python
selected_areas = [
    "World",
    "Northern Africa",
    "Southern Africa",
    "Western Africa",
    "Northern America",
    "Central America",
    "South America",
    "Central Asia",
    "Eastern Asia",
    "Southern Asia",
    "Western Asia",
    "Eastern Europe",
    "Northern Europe",
    "Southern Europe",
    "Western Europe",
    "Oceania",
    "Australia and New Zealand",
]

df_almond_continents = df_almond[df_almond.Area.isin(selected_areas)]
```

```python
df_almond_continents.shape
```

```python
df_almond_continents.head()
```

```python
pd.melt(
    df_almond_continents,
    id_vars=["Area", "Item", "Element", "Unit"],
    value_vars=[col for col in df_almond_continents.columns if col.startswith("Y")],
    var_name="Year",
    value_name="Value",
)
```

```python
# df Small Almonds Continents

df_sac = pd.melt(
    df_almond_continents,
    id_vars=["Area", "Item", "Element", "Unit"],
    value_vars=[col for col in df_almond_continents.columns if col.startswith("Y")],
    var_name="Year",
    value_name="Value",
    ignore_index=True,
)
df_sac.head()
```

```python
df_sac['Measure']=df_sac['Element']+" ("+df_sac['Unit'].fillna('')+")"
df_sac.head()
```

```python
df_sac.drop(columns=['Element', 'Unit'],errors='ignore', inplace=True)
```

```python
df_sac.head()
```

```python
# See if how many items in Item?
df_sac['Item'].nunique()
```

```python
# Drop this column then
df_sac.drop(columns=['Item'], inplace=True)
```

```python
# Clean the Year column to make it integer

df_sac['Year']=df_sac['Year'].str.extract('(\d+)', expand=False)
df_sac.head()
```

```python
df_sac.info()
```

```python
df_sac['Year']=df_sac['Year'].astype(int)
df_sac.info()
```

```python
# Pivting with Value as value, and Measure as column,the rest are rows or index 

df_sacpv = df_sac.pivot_table(index=['Area', 'Year'], columns="Measure", values="Value")
df_sacpv.head()
```

```python
# to repeat items in rows
df_sacpv.reset_index()
```

```python
# make it permanent
df_sacpv.reset_index(inplace=True)
df_sacpv
```
